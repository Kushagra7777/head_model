```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df_GRE = pd.read_csv('GSE210329_miRNA_counts.csv')
df_seto = pd.read_csv('expression_count_test_20240807.csv')
```

```python
# print(df_seto)
df_GRE_1200 = df_GRE.iloc[1200:]
df_GRE_1200 = df_GRE_1200.drop(columns=['ID'])
columns = df_GRE_1200.columns.tolist()
labels = []
for item in columns:
    if item.startswith('BI'):
        labels.append(0)
    elif item.startswith('TO'):
        labels.append(1)    
    else:
        pass    
features = df_GRE_1200['miRNAname'].tolist()
df_ = df_seto[df_seto['gene_id'].isin(features)]
df_ = df_.iloc[:100]
features = df_['gene_id'].tolist()
df_GRE_ = df_GRE_1200[df_GRE_1200['miRNAname'].isin(features)]
#Dropping duplicates
df_GRE_ = df_GRE_.drop_duplicates(subset='miRNAname', keep='first')
df_GRE_.reset_index(drop=True, inplace=True)

#arranging in order
df_GRE_['miRNAname'] = pd.Categorical(df_GRE_['miRNAname'], categories=features, ordered=True)
df_GRE_ = df_GRE_.sort_values(by='miRNAname')

df_GRE_ = df_GRE_.drop(columns=['miRNAname'])
df_GRE_ = df_GRE_.reset_index().drop(columns=['index'])
df_GRE_ = df_GRE_.T

df_ = df_.drop(columns=['gene_id'])
df_ = df_.reset_index().drop(columns=['index'])
df_ = df_.T
# print(df_GRE_)
print(df_)


# df_ = df_seto[df_seto['gene_id'].isin(features)]
# print(df_)
# df_ = df_.iloc[:100]
# print(df_)
# features = df_['gene_id'].tolist()
# print(features)
```
