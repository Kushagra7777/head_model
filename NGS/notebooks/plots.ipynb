```python
import matplotlib.pyplot as plt
import numpy as np

# Sub-models and their correct predictions
sub_models = [
    "random_forest", "logistic_regression", "logistic_regression_sag", 
    "logistic_regression_saga", "extra_tree", "linear_svc", "gbdt", 
    "mlp-3", "mlp-4", "knn-2", "knn-4", "knn-8", "knn-16", "knn-32", 
    "knn-64", "knn-128", "knn-256"
]

# Correct predictions for each model (out of 5 samples)
correct_predictions = [3, 5, 5, 5, 2, 5, 5, 0, 5, 5, 5, 5, 5, 5, 5, 5, 5]

# Incorrect predictions
incorrect_predictions = np.array([5 - correct for correct in correct_predictions])

# Bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the stacked bars
ax.bar(sub_models, correct_predictions, label="Correct Predictions", color='green')
ax.bar(sub_models, incorrect_predictions, bottom=correct_predictions, label="Incorrect Predictions", color='red')

# Adding labels and title
ax.set_ylabel('Number of Predictions')
ax.set_xlabel('Sub-models')
ax.set_title('Sub-model Predictions for 5 Cancer Samples')
plt.xticks(rotation=90)
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = r'E:/MedBank/head_model/NGS/Test_results.xlsx'
df = pd.read_excel(file_path)

# Check and clean up column names if necessary
df.columns = df.columns.str.strip()  # Strip any extra whitespace from headers
print(df.columns)  # Verify the columns after stripping

# Ensure 'Accuracy' column is numeric
df['Accuracy'] = pd.to_numeric(df['Accuracy'], errors='coerce')

# Plotting
plt.figure(figsize=(16, 14))
sns.barplot(x='Accuracy', y='Training_dataset', data=df, hue='Testing_data', palette="viridis")

# Customize plot for readability
plt.title('Model Accuracy by Dataset Combination')
plt.xlabel('Accuracy (%)')
plt.ylabel('Training Dataset')
plt.xlim(0, 1)  # Set limits for percentage scale

# Adjust the legend
plt.legend(title='Testing Data', loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=3, frameon=True)

# Display the plot
plt.show()
```
